# `matplotlib` in ipython

## 畫出圖形存圖

```python
from pylab import figure, plot, savefig
import numpy as np
figure(figsize=(15,10))
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)
plot(x,y)
savefig('matplotlib.png')
```

## 直接畫在ipython notebook上

```python
import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')
```

### 使用[magic function](http://ipython.readthedocs.io/en/stable/interactive/tutorial.html#magics-explained)`%`呼叫`matplotlib` 之中的`inline`後，發現`matplotlib`多出了`style`，可以讓圖形有`ggplot2`的風格。

```python
figure(figsize=(15,10))
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)
plot(x,y)
savefig('matplotlib.png') #依然可以存圖
```

### 不過上緣被切掉一點，調整`ylim`。

```python
from matplotlib.pylab import ylim
```

```python
figure(figsize=(15,10))
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)
z = np.cos(x * np.pi / 180.0)
ylim(-1.1,1.1)
plot(x,y)
plot(x,z) #順手多畫一條cos的線。
```
