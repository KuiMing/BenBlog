# pyecharts

### pyecharts是echarts的python套件，而echarts則是百度公開的javascript工具，用來做資料視覺化。以下以葡萄王（1707）來做示範。

```python
from pandas_datareader import data
from pyecharts import Kline, Line, Overlap, Bar, Grid, Scatter
from datetime import datetime
import numpy as np
from pyecharts import online, Page

online() 
```

### 加上`online()`，jupyter notebook輸出成html，才會能看到圖

## Get Data
### 之前padas可以透過`pandas.io.data`的`web.DataReader`直接拿到股票資料，不過現在改成要透過`pandas_datareader`的`data.DataReader`來取得。`Kline`吃的資料格式是`list`，預設的資料順序是`'Open', 'Close', "High", 'Low'`。

```python
df = data.DataReader('1707.TW', 'yahoo', datetime(2016, 10, 1), datetime.now())
v1 = df[['Open', 'Close', "High", 'Low']].values
```

## K線圖

```python
kline = Kline()
kline.add("日K", df.index, v1)
```

## 加上成交量

```python
bar = Bar()
bar.add("",
    df.index,
    df.Volume,
    is_xaxislabel_align=True,
    legend_top="70%",
    legend_pos="left",
    xaxis_pos="bottom",
)

grid = Grid(width=800, height=600)
grid.add(bar, grid_top="70%")
grid.add(kline, grid_bottom="35%")
```

## 加個移動平均線

```python
line = Line()
line.add("20MA", df.index.astype('str'), df.Close.rolling(window=20).mean())
```

### K線和均線重疊在一起

```python
overlap = Overlap()
overlap.add(kline)
overlap.add(line)
```

### 再加上成交量

```python
grid = Grid(width=800, height=600)
grid.add(bar, grid_top="70%")
grid.add(overlap, grid_bottom="35%")
page = Page()
page.add(grid)
```

## 輸出成html file

```python
page.render('test.html')
```
